### Setup the project schema and volumes and load data in it.

##### Catalog name is `flightsproj`

##### **Raw Data Schema `STAGING`**

```python
%sql
CREATE SCHEMA IF NOT EXISTS flightsproj.staging
```

##### Creating volume `SOURCE` inside the staging schema

```python
%sql
CREATE VOLUME IF NOT EXISTS flightsproj.staging.source
```

Creating individual folders [Flights, Airports, Passengers, Bookings] to store the respected data

```python
# Airports directory
dbutils.fs.mkdirs('/Volumes/flightsproj/staging/source/airports')
```

```python
# Flights directory
dbutils.fs.mkdirs('/Volumes/flightsproj/staging/source/flights')
```

```python
# Passengers directory
dbutils.fs.mkdirs('/Volumes/flightsproj/staging/source/passengers')
```
