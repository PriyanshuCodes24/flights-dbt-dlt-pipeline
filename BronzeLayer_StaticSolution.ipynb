#### **INCREMENTAL DATA INGESTION**  
<small>
Implements idempotent, fault-tolerant ingestion using Databricks Autoloader with Spark Structured Streaming, leveraging checkpointing and the RocksDB state store to provide exactly-once processing semantics.
</small>


### Static Solution

```python
df = spark.readStream.format('cloudFiles')\
        .option('cloudFiles.format', 'csv')\
        .option('cloudFiles.schemaLocation','/Volumes/flightsproj/bronze/bronzevolume/bookings/checkpoint')\
        .option('cloudFiles.schemaEvolutionMode', 'rescue')\
        .load('/Volumes/flightsproj/staging/source/bookings/')
```

```python
df.writeStream.format('delta')\
       .outputMode('append')\
       .trigger(once=True)\
       .option('checkpointLocation','/Volumes/flightsproj/bronze/bronzevolume/bookings/checkpoint')\
       .option('path','/Volumes/flightsproj/bronze/bronzevolume/bookings/data')\
       .start()
        
```
