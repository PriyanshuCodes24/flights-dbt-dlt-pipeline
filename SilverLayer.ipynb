Read and understand the data from bronze layer

```python
df = spark.read.format('delta')\
          .load('/Volumes/flightsproj/bronze/bronzevolume/bookings/data')
```

```python
display(df)
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

Converting the amount column from string to double

```python
df = df.withColumn('amount', col('amount').cast('double'))\
     .withColumn('booking_date', to_date(col('booking_date')))

df.printSchema()
```

```python
df = df.drop("_rescued_data")
```

```python
df.printSchema()
```

```python
# Setting the processed timestamp
df = df.withColumn('modified_date', current_timestamp())
```

```python
display(df)
```
