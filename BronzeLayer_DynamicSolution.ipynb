#### **INCREMENTAL DATA INGESTION**  

### Dynamic Solution

```python
dbutils.widgets.text('src','')
```

```python
src_value = dbutils.widgets.get('src')
```

```python
src_value
```

```python
df = spark.readStream.format('cloudFiles')\
        .option('cloudFiles.format', 'csv')\
        .option('cloudFiles.schemaLocation',f'/Volumes/flightsproj/bronze/bronzevolume/{src_value}/checkpoint')\
        .option('cloudFiles.schemaEvolutionMode', 'rescue')\
        .load(f'/Volumes/flightsproj/staging/source/{src_value}/')
```

```python
df.writeStream.format('delta')\
       .outputMode('append')\
       .trigger(once=True)\
       .option('checkpointLocation',f'/Volumes/flightsproj/bronze/bronzevolume/{src_value}/checkpoint')\
       .option('path',f'/Volumes/flightsproj/bronze/bronzevolume/{src_value}/data')\
       .start()
        
```
